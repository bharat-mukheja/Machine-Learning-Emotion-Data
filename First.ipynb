```python
import pandas as pd
import numpy as np
```

```python
facet1 = pd.read_csv("DataSample/164_FACET_1.txt",delimiter='\t')
facet1.shape
```

```python
All = facet1.ix[:,13:-3]
```

```python
#Removing the rows with empty rows
All = All[pd.notnull(All['FaceRect X'])]
All.shape
```

```python
All.head()
```

```python
X = pd.concat([All.ix[:,:6],All.ix[:,30:]],axis=1)
X.head()
```

```python
Y = All.ix[:,6:30]
Y.head()
```

```python
yJoyE = Y.ix[:,0]
yJoyE.head()
```

```python
import xgboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,explained_variance_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, yJoyE, test_size=0.35, random_state=123)
```

```python
xg_model = xgboost.XGBRegressor(n_estimators=100, learning_rate=0.1, gamma=0.1, subsample=1, colsample_bytree=1, max_depth=5)
xg_model.fit(X_train, y_train)
y_xg_predict = xg_model.predict(X_test)
y_train_predict=xg_model.predict(X_train)
mse=mean_squared_error(y_test, y_xg_predict)
variation = explained_variance_score(y_test,y_xg_predict)
mse_training=mean_squared_error(y_train, y_train_predict)
print("*********")
print("MSE Validation:"+str(mse))
print("MSE Training:"+str(mse_training))
print("Explained variation:"+str(variation))
print("*********")
```
